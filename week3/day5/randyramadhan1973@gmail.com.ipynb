```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('https://blog.sanbercode.com/wp-content/uploads/2020/05/vgsales.csv')
df.head()
```

<h2>Quiz 1 Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
GenreMean = df.groupby('Genre')[['NA_Sales','EU_Sales','JP_Sales', 'Other_Sales']].mean()
GenreMean
```

Expected Output :

![image.png](https://blog.sanbercode.com/wp-content/uploads/2020/05/image-8.png)

<h2>Perbandingan Kuantitatif Barplot : Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
df = GenreMean.T
df
```

```python
df.plot.bar(figsize=(17,8), rot=0, width=0.8)
plt.style.use('ggplot')

plt.title('Mean Sales Video Game By Genre')
plt.show()
```

insight yang bisa dapat dari visualisasi tersebut:
1. penggunaan Transpose pada dataframe
2. penggunaan bar secara ringkas

<h2>Perbandingan Kuantitatif Barplot : Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
df.plot.bar(figsize=(17,8), rot=0, width=0.8, stacked=True)
plt.style.use('ggplot')

plt.title('Mean Sales Video Game By Genre')
plt.show()
```
