<h2>Tugas Harian No 2 : Membuat Component Figure dan Axis</h2>

- Buatlah Figure dan Axis kosong tanpa Data
- Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y = np.cos(x/2)
```

```python
fg = plt.figure()
fg.add_subplot()
plt.show()
```

Exected Output :

![image.png](https://blog.sanbercode.com/wp-content/uploads/2020/05/image-3.png)

```python
# implicit
# membuat objek figure
fig = plt.figure()

# membuat objek axes
ax = fig.add_subplot()

# menambahkan dalam ke objek axes
ax.plot(x, y)
# menampilkan data
plt.show()
```

```python
# explicit
plt.plot(x,y)
plt.show()
```

<h2>Tugas Harian No 3 : Memasukan multiple data ke dalam satu axis</h2>

- Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style
- save data tersebut dengan nama file bebas

```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)
```

```python
plt.style.use("seaborn")
plt.subplot()
plt.plot(x,y)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```
