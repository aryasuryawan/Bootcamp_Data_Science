```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>No 1 : Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output :

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
fig, ax = plt.subplots(ncols=3, figsize=(15,6), sharey=True, sharex=True, )
ax[0].plot(x, tan, color='B', label='tanX')
ax[0].set_title('Grafik tanx')
ax[1].plot(x, cos, color='R', label='sinX')
ax[1].set_title('Grafik sinx')
ax[2].plot(x, sin, color='G', label='cosX') 
ax[2].set_title('Grafik cosx')

fig.legend(loc='upper center')
plt.show()
```

<h2>No 2 : Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output :

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
fig, ax = plt.subplots(figsize=(10,5))

ax.plot(x,y)
ax.set_title('Grafik CosX')

# aka
aka = ax.inset_axes([0.9,0.8,0.3,0.4])
aka.plot(x, y)
aka.plot(x, y4, color='#263238')
aka.set_title('Grafik CosX^4')


#aki
aki = ax.inset_axes([-0.2,0.8,0.3,0.4])
aki.plot(x, y)
aki.plot(x, y5, color='y')
aki.set_title('Grafik CosX^5')


#bki
bki = ax.inset_axes([-0.2,-0.2,0.3,0.4])
bki.plot(x, y)
bki.plot(x, y2, color='r')
bki.set_title('Grafik CosX^2')


#bka
bka = ax.inset_axes([0.9,-0.2,0.3,0.4])
bka.plot(x, y)
bka.plot(x, y3, color='g')
bka.set_title('Grafik CosX^3')
```
